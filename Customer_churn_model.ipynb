```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df.dropna()
```

```python
#B) Data Visulization : 
#a) Build  a pie chart to show the distribution of customers would be churning out
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    labels=['No','Yes']
)
plt.title("Customer Churn Distribution")
plt.ylabel('')
plt.show()
```

```python
#b) Build a bar plot to show the distribution of 'Internet service'
sns.countplot(x='InternetService',data=df)
plt.title("Internet Service Distribution")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x=df[['tenure']]
y=df[['Churn']]
scaler=StandardScaler()
x=scaler.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential()

# Input Layer
model.add(Input(shape=(1,)))

# Hidden Layers
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))

# Output Layer
model.add(Dense(1, activation='sigmoid'))

# Compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
hist1 = model.fit(x_train, y_train, epochs=150, validation_data=(x_test, y_test))
```

```python
np.mean(hist1.history['val_accuracy'])
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
plt.plot(hist1.history['accuracy'])
plt.plot(hist1.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train','Validation'])
plt.show()
```

```python
#B) model
from tensorflow.keras.layers import Dropout
model2=Sequential()
#Input Layer
model2.add(Dense(12,input_dim=1,activation='relu'))
model2.add(Dropout(0.3))

#Hidden Layer
model2.add(Dense(8,activation='relu'))
model2.add(Dropout(0.2))

#Output Layer
model2.add(Dense(1,activation='sigmoid'))

model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

hist2=model2.fit(x_train,y_train,epochs=150,validation_data=(x_test,y_test))
```

```python

y_pred2=(model2.predict(x_test)>0.5).astype(int)


#Confusion Matrix
print(confusion_matrix(y_test,y_pred2))
```

```python
#Accuracy Graph
plt.plot(hist2.history['accuracy'])
plt.plot(hist2.history['val_accuracy'])
plt.title('Model 2 Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['Train','Test'])
plt.show()
```

```python
#model 3 (3 features)
#Split  data 
x=df[['tenure','MonthlyCharges','TotalCharges']]
y=df[['Churn']]

x=scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

```

```python
#Build Model
model3=Sequential()

model3.add(Dense(12,activation='relu',input_dim=3))
model3.add(Dense(8,activation='relu'))
model3.add(Dense(1,activation='sigmoid'))


model3.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

hist3=model3.fit(x_train,y_train,epochs=150,validation_data=(x_test,y_test))

    
```

```python
y_pred3=(model3.predict(x_test)>0.5).astype(int)
print(confusion_matrix(y_test,y_pred3))
```
